```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
# sampling rate
sr = 100.0
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

# frequency of the signal
freq = 1
y = np.sin(2*np.pi*freq*t)

plt.figure(figsize = (4, 3))
plt.subplot(211)
plt.plot(t, y, 'b')

freq = 3
y = np.sin(2*np.pi*freq*t)

plt.figure(figsize = (4, 3))
plt.subplot(211)
plt.plot(t, y, 'b')

freq = 6
y = np.sin(2*np.pi*freq*t)

plt.figure(figsize = (4, 3))
plt.subplot(211)
plt.plot(t, y, 'b')


freq = 9
y = np.sin(2*np.pi*freq*t)

plt.figure(figsize = (4, 3))
plt.subplot(212)
plt.plot(t, y, 'b')
plt.xlabel('Position (w)')
plt.show()
```
